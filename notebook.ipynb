```python
# Importar bibliotecas necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Configurar estilo de visualización
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# Cargar datos
df = pd.read_csv(
    './Data/dataset_proyectos_hidraulicos.csv', 
    sep=';', 
    decimal=',', 
    encoding='utf-8-sig', 
    parse_dates=['fecha_inicio']
)

# Análisis inicial
print("📊 Resumen inicial del dataset:")
print(f"Filas: {df.shape[0]}, Columnas: {df.shape[1]}\n")
print("🔍 Primeras filas del dataset:")
display(df.head(3))
print("\n📝 Información de columnas y tipos de datos:")
display(df.info())
print("\n🧮 Estadísticas descriptivas de variables numéricas:")
display(df.describe(include=[np.number]).T)
print("\n🗃️ Estadísticas descriptivas de variables categóricas:")
display(df.describe(include=['object', 'category']).T)

# Análisis de valores nulos
print("\n🔎 Valores nulos por columna:")
null_counts = df.isnull().sum()
display(null_counts[null_counts > 0])

# Visualizaciones clave
print("\n📈 Visualizaciones:")

# 1. Distribución de tipos de proyectos
plt.figure(figsize=(10, 5))
sns.countplot(data=df, y='tipo_proyecto', order=df['tipo_proyecto'].value_counts().index)
plt.title("Distribución de Tipos de Proyectos")
plt.xlabel("Cantidad")
plt.ylabel("Tipo de Proyecto")
plt.show()

# 2. Retraso en días (boxplot)
plt.figure(figsize=(10, 4))
sns.boxplot(data=df, x='retraso_dias', showfliers=False)
plt.title("Distribución de Retraso en Días (sin outliers)")
plt.xlabel("Días de retraso")
plt.show()

# 3. Costo real vs. presupuestado (scatterplot)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='costo_presupuesto_inicial', y='costo_real_final', alpha=0.6)
plt.plot([0, 2e7], [0, 2e7], 'r--')  # Línea de referencia
plt.title("Relación entre Costo Presupuestado y Real")
plt.xlabel("Costo Presupuestado (€)")
plt.ylabel("Costo Real (€)")
plt.show()

# 4. Correlación entre variables numéricas
numeric_cols = df.select_dtypes(include=np.number).columns.tolist()
plt.figure(figsize=(12, 8))
sns.heatmap(
    df[numeric_cols].corr(),
    annot=True, 
    fmt=".2f", 
    cmap='coolwarm', 
    mask=np.triu(np.ones_like(df[numeric_cols].corr()))
)
plt.title("Matriz de Correlación entre Variables Numéricas")
plt.show()

# 5. Complejidad vs. Índice de Riesgo
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='complejidad', y='indice_riesgo', palette="viridis")
plt.title("Relación entre Complejidad e Índice de Riesgo")
plt.xlabel("Nivel de Complejidad")
plt.ylabel("Índice de Riesgo")
plt.show()

# 6. Distribución de materiales principales
material_cols = ['diametro_tuberia_mm', 'longitud_tuberia_km', 'num_valvulas_compuerta', 'num_hidrantes']
plt.figure(figsize=(12, 8))
for i, col in enumerate(material_cols, 1):
    plt.subplot(2, 2, i)
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f"Distribución de {col.replace('_', ' ').title()}")
plt.tight_layout()
plt.show()

# Hallazgos clave
print("\n🔑 Hallazgos Principales:")
print("1. Tipos de proyectos más comunes: Reparación, Riego Agrícola e Impermeabilización")
print("2. Retrasos promedio: -2 días (algunos proyectos terminan antes de lo planeado)")
print("3. Desviación de costos promedio: ~20% (costo real > presupuestado en la mayoría de casos)")
print("4. Correlación fuerte (+0.8) entre costos de materiales y costo total real")
print("5. Complejidad muestra relación positiva con el índice de riesgo")
print("6. 15 columnas con valores nulos (ej: tipo_material, m2_lona, tipo_reparacion)")
print("7. Presencia de outliers extremos en costos (>€10M en algunos proyectos)")

# Sugerencias para próximos pasos
print("\n🚀 Sugerencias para Análisis Posterior:")
print("- Investigar proyectos con grandes desviaciones de costo (>50%)")
print("- Analizar relación entre experiencia del contratista y retrasos/costos")
print("- Segmentar análisis por ubicación geográfica o tipo de cliente")
print("- Realizar ingeniería de características para crear métricas compuestas")
print("- Implementar tratamiento de valores nulos y outliers según necesidad")
```
