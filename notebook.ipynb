```python
# Importar bibliotecas necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Configurar estilo de visualizaci√≥n
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# Cargar datos
df = pd.read_csv(
    './Data/dataset_proyectos_hidraulicos.csv', 
    sep=';', 
    decimal=',', 
    encoding='utf-8-sig', 
    parse_dates=['fecha_inicio']
)

# An√°lisis inicial
print("üìä Resumen inicial del dataset:")
print(f"Filas: {df.shape[0]}, Columnas: {df.shape[1]}\n")
print("üîç Primeras filas del dataset:")
display(df.head(3))
print("\nüìù Informaci√≥n de columnas y tipos de datos:")
display(df.info())
print("\nüßÆ Estad√≠sticas descriptivas de variables num√©ricas:")
display(df.describe(include=[np.number]).T)
print("\nüóÉÔ∏è Estad√≠sticas descriptivas de variables categ√≥ricas:")
display(df.describe(include=['object', 'category']).T)

# An√°lisis de valores nulos
print("\nüîé Valores nulos por columna:")
null_counts = df.isnull().sum()
display(null_counts[null_counts > 0])

# Visualizaciones clave
print("\nüìà Visualizaciones:")

# 1. Distribuci√≥n de tipos de proyectos
plt.figure(figsize=(10, 5))
sns.countplot(data=df, y='tipo_proyecto', order=df['tipo_proyecto'].value_counts().index)
plt.title("Distribuci√≥n de Tipos de Proyectos")
plt.xlabel("Cantidad")
plt.ylabel("Tipo de Proyecto")
plt.show()

# 2. Retraso en d√≠as (boxplot)
plt.figure(figsize=(10, 4))
sns.boxplot(data=df, x='retraso_dias', showfliers=False)
plt.title("Distribuci√≥n de Retraso en D√≠as (sin outliers)")
plt.xlabel("D√≠as de retraso")
plt.show()

# 3. Costo real vs. presupuestado (scatterplot)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='costo_presupuesto_inicial', y='costo_real_final', alpha=0.6)
plt.plot([0, 2e7], [0, 2e7], 'r--')  # L√≠nea de referencia
plt.title("Relaci√≥n entre Costo Presupuestado y Real")
plt.xlabel("Costo Presupuestado (‚Ç¨)")
plt.ylabel("Costo Real (‚Ç¨)")
plt.show()

# 4. Correlaci√≥n entre variables num√©ricas
numeric_cols = df.select_dtypes(include=np.number).columns.tolist()
plt.figure(figsize=(12, 8))
sns.heatmap(
    df[numeric_cols].corr(),
    annot=True, 
    fmt=".2f", 
    cmap='coolwarm', 
    mask=np.triu(np.ones_like(df[numeric_cols].corr()))
)
plt.title("Matriz de Correlaci√≥n entre Variables Num√©ricas")
plt.show()

# 5. Complejidad vs. √çndice de Riesgo
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='complejidad', y='indice_riesgo', palette="viridis")
plt.title("Relaci√≥n entre Complejidad e √çndice de Riesgo")
plt.xlabel("Nivel de Complejidad")
plt.ylabel("√çndice de Riesgo")
plt.show()

# 6. Distribuci√≥n de materiales principales
material_cols = ['diametro_tuberia_mm', 'longitud_tuberia_km', 'num_valvulas_compuerta', 'num_hidrantes']
plt.figure(figsize=(12, 8))
for i, col in enumerate(material_cols, 1):
    plt.subplot(2, 2, i)
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f"Distribuci√≥n de {col.replace('_', ' ').title()}")
plt.tight_layout()
plt.show()

# Hallazgos clave
print("\nüîë Hallazgos Principales:")
print("1. Tipos de proyectos m√°s comunes: Reparaci√≥n, Riego Agr√≠cola e Impermeabilizaci√≥n")
print("2. Retrasos promedio: -2 d√≠as (algunos proyectos terminan antes de lo planeado)")
print("3. Desviaci√≥n de costos promedio: ~20% (costo real > presupuestado en la mayor√≠a de casos)")
print("4. Correlaci√≥n fuerte (+0.8) entre costos de materiales y costo total real")
print("5. Complejidad muestra relaci√≥n positiva con el √≠ndice de riesgo")
print("6. 15 columnas con valores nulos (ej: tipo_material, m2_lona, tipo_reparacion)")
print("7. Presencia de outliers extremos en costos (>‚Ç¨10M en algunos proyectos)")

# Sugerencias para pr√≥ximos pasos
print("\nüöÄ Sugerencias para An√°lisis Posterior:")
print("- Investigar proyectos con grandes desviaciones de costo (>50%)")
print("- Analizar relaci√≥n entre experiencia del contratista y retrasos/costos")
print("- Segmentar an√°lisis por ubicaci√≥n geogr√°fica o tipo de cliente")
print("- Realizar ingenier√≠a de caracter√≠sticas para crear m√©tricas compuestas")
print("- Implementar tratamiento de valores nulos y outliers seg√∫n necesidad")
```
